```python
import os
from pdf2image import convert_from_path
from PIL import Image
import re
import pytesseract

pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

TabelentryAIL = ""
whitelist = r'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789.: |\-'
custom_config = r'preserve_interword_spaces=1 --psm 6 --oem 1 -c tessedit_char_whitelist=' + whitelist
awal = 1
akhir = 500
tipedok = ["", "PK", "BA", "SPJBTL", "OTHER", "1", "2", "3", "4", "5", "6", "7", "8", "8", "10", "11", "12"] + list(range(13, 25))
DIR = os.getcwd()

while awal < akhir + 1:
    l = str(awal)
    pages = []
    pdf_found = False
    
    for k in tipedok:
        for m in [""] + list(range(1, 10)):
            n = f"_{m}" if m != "" else ""
            pdfs = os.path.join(DIR, f"{l}-{k}{n}.pdf")
            
            if os.path.exists(pdfs):
                pdf_found = True
                try:
                    read = convert_from_path(
                        pdfs,
                        poppler_path=r"C:\ProgramData\anaconda3\envs\AIL\Library\bin",
                        fmt="jpeg",
                        jpegopt={'optimize': True, 'progressive': False, 'quality': 50},
                        size=(2480, 3508)
                    )
                except (OSError, Image.UnidentifiedImageError) as e:
                    print(f"Error processing image: {pdfs}\nSkipping...")
                    continue
                
                for r in read:
                    convert = r.convert('P', palette=Image.ADAPTIVE)
                    pages.append(convert)
    
    if not pdf_found:
        awal += 1
        continue
    
    text = ""
    for j, page in enumerate(pages):
        text += str(pytesseract.image_to_string(page, config='AIL --psm 1 --oem 1'))
    
    patternNama = r"Nama[\-\:\.\_\s\|]+(?:P?(?=[elangq ]))[elangq ]*[\-\:\.\_\s\|]*([A-Z\.\d\_ ]+(?![a-z ]))"
    patternNoPel = r"[NoID]+[\.\s\-\_\|]+Pelan[cagq]+[ae]+n[\\_\.\s\-\:\|]+(\d+)"
    patternNoPel2 = r"\n?\b(5613\d{8}|53951\d{7})\b"
    patternNoAg = r"[Nomr\.]+[\.\s\-\:\_\|\,\*]+A[gaq]+enda[\\_\.\s\-\:\|\,\*]+(\d+)"
    patternNoAg2 = r"KEDUA[\\_\.\s\-\:\|]+Nomor[\\_\.\s\-\:\|]+(\d+)"
    patternNoAg3 = r"\n?\b(5613\d{14}|53951\d{13})\b"
    
    print(l)
    NamaPelsearch = re.search(patternNama, text)
    NamaPel = NamaPelsearch.group(1) if NamaPelsearch else ""
    print(NamaPel)
    
    NoPellist = ["", ""]
    NoPelsearch = re.search(patternNoPel, text)
    NoPelsearch2 = re.search(patternNoPel2, text)
    NoPellist[0] = NoPelsearch.group(1) if NoPelsearch else ""
    NoPellist[1] = NoPelsearch2.group(1) if NoPelsearch2 else ""
    NoPel = ",".join(NoPellist)
    print(NoPel)
    
    NoAglist = []
    NoAgsearch = re.search(patternNoAg, text)
    NoAgsearch2 = re.search(patternNoAg2, text)
    NoAgsearch3 = re.search(patternNoAg3, text)
    NoAglist.append(NoAgsearch.group(1) if NoAgsearch else "")
    NoAglist.append(NoAgsearch2.group(1) if NoAgsearch2 else "")
    NoAglist.append(NoAgsearch3.group(1) if NoAgsearch3 else "")
    NoAg = ",".join(NoAglist)
    print(NoAg)
    
    filecsv = os.path.join(DIR, "TabelentryAILsplit2.csv")
    entryAIL = f"{l},{NamaPel},,{NoPel},{NoAg}\n"
    with open(filecsv, "a") as tulisTabel:
        tulisTabel.write(entryAIL)
    
    TabelentryAIL += entryAIL
    awal += 1
    
print(TabelentryAIL)
```
